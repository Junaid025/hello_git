# Range Normalization

```python
df['price'].describe()
Min = 2.004660e+05
price = df['price'].to_numpy()
price = price-Min
Max = max(price)
price = price/Max
price
```

```python
df['parking'].describe()
Min = 0
parking = df['parking'].to_numpy()
parking = parking-Min
Max = max(parking)
parking = parking/Max
parking
```

```python
df['stories'].describe()
Min = 1
stories = df['stories'].to_numpy()
stories = stories-Min
Max = max(stories)
stories = stories/Max
stories
```

```python
df['bathrooms'].describe()
Min = 1
bathrooms = df['bathrooms'].to_numpy()
bathrooms = bathrooms-Min
Max = max(bathrooms)
bathrooms = bathrooms/Max
bathrooms
```

```python
df['bedrooms'].describe()
Min = 1
bedrooms = df['bedrooms'].to_numpy()
bedrooms = bedrooms-Min
Max = max(bedrooms)
bedrooms = bedrooms/Max
bedrooms
```

```python
df['area'].describe()
Min = 94828.000000
Area = df['area'].to_numpy()
Area = Area-Min
Max = max(Area)
Area = Area/Max
Area
```

```python
df['length'].describe()
Min = 300.000000
length = df['length'].to_numpy()
length = length-Min
Max = max(length)
length = length/Max
length
```

```python
df['width'].describe()
Min = 300.000000
width = df['width'].to_numpy()
width = width-Min
Max = max(width)
width = width/Max
width
```

```python
df['furnishingstatus'].replace("furnished",1 ,inplace = True)
df['furnishingstatus'].replace("unfurnished",0 ,inplace = True)
df['furnishingstatus'].replace("semi-furnished",-1 ,inplace = True)
```

```python
df
```

```python
y = df["furnishingstatus"]
df = df.drop(['mainroad' ,'guestroom' ,'basement', 'hotwaterheating' ,'airconditioning','prefarea','furnishingstatus'],axis = 1)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test =train_test_split(df, y, test_size=0.4, random_state=50)

clf1=KNeighborsClassifier()
clf1.fit(X_train, y_train)
pred=clf1.predict(X_test)
 
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("F1 Score : ",f1_score(y_test, pred, average='micro')*100)
```

# Covarience and Corelation

```python
covar  = normalized.to_numpy()
trans =  covar.transpose()
covar  = np.dot(trans,covar)
print(covar)
```

```python
covar.shape
```

```python
price_std = normalized["price"].std()
length_std = normalized["length"].std()
width_std = normalized["width"].std()
bedrooms_std = normalized["bedrooms"].std()
bathrooms_std = normalized["bathrooms"].std()
stories_std = normalized["stories"].std()
parking_std = normalized["parking"].std()
Area_std = normalized["area"].std()
net_std = price_std*length_std*width_std*bedrooms_std*bathrooms_std*stories_std*parking_std*Area_std
corelation = covar/net_std
corelation
```

# Corelation of lenght and Area

```python
covar =  normalized[["area","length"]]
covar =  covar.to_numpy()
trans =  covar.transpose()
covar  = np.dot(trans,covar)
co_relation = covar/(length_std*Area_std)
co_relation
```
